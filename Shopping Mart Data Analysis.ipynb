## Qs 1 .What was the best month for sales ? How much was earned that month? 

Process:
1. Covert quanty & Price Each into numberic value 
2. Multiple Quanty & Price Each to get total for each sale 
3. Group by total sales for each month 

```python
df_filtered["Quantity Ordered"] = df_filtered["Quantity Ordered"].astype(float)
df_filtered["Price Each"] = df_filtered["Price Each"].astype(float)

```

```python
df_filtered["Sale"] = df_filtered["Quantity Ordered"]*df_filtered["Price Each"]
```

```python
df_filtered.head(2)
```

```python
df = df_filtered.groupby("Month")["Sale"].sum().sort_values(ascending=False)
df = df.to_frame()
df
```

```python
list = ['December','October','April','November','May','March','July','June','August','February','September','January']
df['month'] = list
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize = ( 11.7 , 7.27 ))
sns.barplot(y='month',x='Sale',data=df)
ax.set_title( "Month wise Sales" , size = 14 )
ax.set_xlabel("Sales * 1000000")
plt.show()
```

Best Month for selling is December & Total value of  $ 4619297

# Qs 2 . Which city had the highest number of sales ?




Process:
1. Get City from Purchase Address
2. Group by each city, sum up total sales 

```python
city= []
df_temp = df_filtered

for i in df_filtered["Purchase Address"]:
    city.append(i.split(",")[1])
df_temp["city"] = city
df_temp.head()
```

```python
df_city = df_temp.groupby('city')['Sale'].sum().sort_values(ascending=False)
df_city = df_city.to_frame()
df_city
```

```python
list =['San Francisco','Los Angeles','New York City','Boston','Atlanta','Dallas','Seattle','Portland','Austin']
df_city['city'] = list
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize = ( 11.7 , 7.27 ))
sns.barplot(y='city',x='Sale',data=df_city)
ax.set_title( "City wise Sales" , size = 14 )
ax.set_xlabel("Sales * 1000000")
plt.show()
```

San Francisco got the most sales $ 8262203

### Qs 3. What products sold the most ?

Process:
1. Group by product and sum up Quantity order

```python
df_product = df_filtered.groupby('Product')['Quantity Ordered'].sum().sort_values(ascending=False).head().to_frame()
df_product
```

```python
list = ['AAA Batteries (4-pack)','AA Batteries (4-pack)','USB-C Charging Cable','Lightning Charging Cable','Wired Headphones']
df_product['Product'] = list
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize = ( 11.7 , 7.27 ))
sns.barplot(y='Product',x='Quantity Ordered',data=df_product)
ax.set_title( "Most Sold Prodcuts" , size = 14 )
plt.show()
```

AAA Batteries (4-pack) got sold the most. Quantity of the order: 31017 & its cost the minimum from all products
